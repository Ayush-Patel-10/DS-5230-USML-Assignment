```python
import numpy as np
import matplotlib.pyplot as plt

mean = np.array([0, 0])
cov = np.array([[1, 0.5], [0.5, 1]]) 

def sample_x1_given_x2(x2):
    #Calculating the conditional mean and covariance for x1 given x2
    conditional_mean = mean[0] + cov[0, 1] / cov[1, 1] * (x2 - mean[1])
    conditional_cov = cov[0, 0] - cov[0, 1] / cov[1, 1] * cov[1, 0]
    
    #Get sample from x1 given x2
    x1 = np.random.normal(conditional_mean, np.sqrt(conditional_cov))
    
    return x1

def sample_x2_given_x1(x1):
    #Calculating the conditional mean and covariance for x2 given x1
    conditional_mean = mean[1] + cov[1, 0] / cov[0, 0] * (x1 - mean[0])
    conditional_cov = cov[1, 1] - cov[1, 0] / cov[0, 0] * cov[0, 1]
    
    #Get sample from x2 given x1
    x2 = np.random.normal(conditional_mean, np.sqrt(conditional_cov))
    
    return x2

x_result = 0  
y_result = 0  

#Number of iterations
num_iterations = 1000

x = np.zeros(num_iterations)
y = np.zeros(num_iterations)

for i in range(num_iterations):

    #Sampling from x1 given x2
    x_result = sample_x1_given_x2(x_result)
    #Sampling from x2 given x1
    y_result = sample_x2_given_x1(y_result)
    
    # Store the samples
    x[i] = x_result
    y[i] = y_result

plt.plot(x, y, '*')
plt.show()
```
